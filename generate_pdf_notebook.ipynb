# Generate a PDF from Markdown in a Jupyter Notebook
This notebook demonstrates how to install required libraries, prepare the conversion workflow, and export a markdown file to PDF. It will also verify the generated PDF file.

## 2. Create Notebook Content
The notebook will use the markdown file `CLIENT_READY_RECKONER.md` as the source content and convert it into a PDF file named `CLIENT_READY_RECKONER.pdf`.

```python
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
import re

source_md = 'CLIENT_READY_RECKONER.md'
output_pdf = 'CLIENT_READY_RECKONER.pdf'

styles = getSampleStyleSheet()
styles['Heading1'].fontSize = 18
styles['Heading1'].leading = 22
styles['Heading1'].spaceAfter = 12
styles['Heading2'].fontSize = 14
styles['Heading2'].leading = 18
styles['Heading2'].spaceAfter = 10
styles['BodyText'].fontSize = 11
styles['BodyText'].leading = 14
styles['BodyText'].spaceAfter = 6
if 'Bullet' not in styles.byName:
    styles.add(ParagraphStyle(name='Bullet', parent=styles['BodyText'], leftIndent=18, bulletIndent=9, spaceBefore=0, spaceAfter=0))

with open(source_md, 'r', encoding='utf-8') as f:
    lines = [line.rstrip('\n') for line in f]

story = []
for line in lines:
    if not line.strip():
        story.append(Spacer(1, 6))
        continue
    if line.startswith('# '):
        story.append(Paragraph(line[2:].strip(), styles['Heading1']))
        continue
    if line.startswith('## '):
        story.append(Paragraph(line[3:].strip(), styles['Heading2']))
        continue
    if line.startswith('### '):
        story.append(Paragraph(line[4:].strip(), styles['Heading2']))
        continue
    m = re.match(r'^(\s*)[-*+]\s+(.*)', line)
    if m:
        story.append(Paragraph(m.group(2), styles['Bullet'], bulletText='•'))
        continue
    story.append(Paragraph(line, styles['BodyText']))


doc = SimpleDocTemplate(output_pdf, pagesize=letter, rightMargin=40, leftMargin=40, topMargin=40, bottomMargin=40)
doc.build(story)
print(f'Created {output_pdf}')
```

## 3. Configure PDF Conversion
We configure the PDF output using ReportLab styles and the document template. This setup controls margins, fonts, heading sizes, and bullet formatting.

## 4. Export Notebook to PDF
Run the conversion code cell above to export the markdown file into a PDF document. The created PDF will appear in the workspace alongside this notebook.
